```python
def preprocess(text):
    text = text.lower()
    text = re.sub(r'[^а-яёa-z\s]', ' ', text)
    text = re.sub(r'\s+', ' ', text).strip()
    stop_words = set(stopwords.words('russian'))
    words = [w for w in text.split() if w not in stop_words]
    return ' '.join(words)

clean_texts = [preprocess(t) for t in texts]
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(clean_texts)
sim_matrix = cosine_similarity(tfidf_matrix)
df_sim = pd.DataFrame(sim_matrix, index=titles, columns=titles)
print("Косинусное сходство между текстами:")
print(df_sim.round(3))
```

Этот прогноз подтвердился: пара Текст 2 и Текст 4 получила максимальное косинусное сходство 0.597. Второе место заняла пара Текст 2 и Текст 3 – такой результат уже менее очевиден, ведь Текст 3 лирически описывает быт пекарей на островах Силли и далёк от макроэкономики. Удивило и то, что Текст 1 («менеджмент») оказался наиболее близок именно к «пекарне» (Текст 3, 0.520), а не к деловым текстам 2 и 4 (с ними сходство всего 0.482 и 0.477).

Причина такого распределения – особенность TF-IDF векторизации на униграммах. Удалив стоп‑слова, мы всё равно оставляем широкий пласт общеупотребительной лексики, связанной с работой и бизнесом: work, business, time, company, shop, provide. В «менеджменте» и «пекарне» эти слова повторяются регулярно (планирование, работа, магазин, обслуживание), что создаёт иллюзию тематической близости. Тексты 2 и 4, напротив, перенасыщены специализированной терминологией (shipping, bunker fuel, government, FDI, Dunning, internalisation), которая почти не пересекается с лексикой общего менеджмента из Текста 1, отсюда и низкие значения. TF-IDF фиксирует не смысловую, а чисто лексическую общность, и в данном корпусе именно повседневная «рабочая» лексика оказалась сильным связующим фактором, иногда перевешивающим тематическое сходство.
